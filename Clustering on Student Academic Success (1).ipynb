# Clustering on Student Academic Success

#### In this part of the notebook, we investigate the profiles of successful students via clustering

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics

studentData = pd.read_csv('dataset.csv')
studentData.head()
```

### Exploratory Data Analysis

```python
studentData.info()
```

### Courses


```python
# Does being in different courses affect the dropout rate (success rate)
sb.heatmap(studentData.groupby(['Course', 'Target']).size().unstack(), 
           linewidths = 1, annot = True, fmt = 'g', annot_kws = {"size": 10}, cmap = "BuGn")
```

### Most of the Nursing students are Graduates

```python
sb.heatmap(studentData.groupby(['Scholarship holder', 'Target']).size().unstack(), 
           linewidths = 1, annot = True, fmt = 'g', annot_kws = {"size": 18}, cmap = "BuGn")
```

### Out of 4424 students, 24.8% of the students are scholarship holders. Out of the Scholarship Holders, 12.2% of the students are Dropouts.
